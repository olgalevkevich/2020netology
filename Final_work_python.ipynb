```python
import pandas as pd
import datetime
```

  1. Загрузите файлы по оценкам (ratings) и фильмам (movies) и создайте на их основе pandas-датафреймы

```python
Ratings = pd.read_csv('u.data.csv')
Ratings.head()
```

```python
Ratings.info()
```

```python
#переименую item_id в movie_id для дальнейшего соединения датафреймов
Ratings.rename(columns={'item_id': 'movie_id'}, inplace=True)
Ratings.head()
```

```python
movies = pd.read_csv('u.item.csv', sep="|", encoding='latin-1', header=None)
movies.columns = ['movie_id', 'movie_title', 'release_date', 'video_release date', 'IMDb_URL', 'unknown', 'Action', 'Adventure', 'Animation', 'Childrens', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']
movies.head()
```

```python
movies.info()
```

  2. Средствами Pandas, используя dataframe ratings, найдите id пользователя, поставившего больше всего оценок

```python
count_rating = Ratings['user_id'].value_counts().sort_values(ascending=False)
count_rating.head(1)
```

  3. Оставьте в датафрейме ratings только те фильмы, который оценил данный пользователь 

```python
filtered_films = Ratings[Ratings['user_id'] == 405]
filtered_films.head()
```

```python
#добавляю столбец с годом выхода
movies['release_year'] = pd.DatetimeIndex(movies['release_date']).year
movies.head()
```

```python
#считаю количестто оценок и суммарную оценку по каждому фильму
ratigs_group = Ratings.groupby('movie_id').agg({'user_id': 'count', 'rating': 'sum'}).reset_index()
ratigs_group.rename(columns={'user_id': 'rating_count', 'rating': 'rating_sum'}, inplace=True)
ratigs_group.head()
```

  4.1. Добавьте к датафрейму из задания 3 столбцы по жанрам. Каждый столбец - это жанр. Единицу записываем, если фильм принадлежит данному жанру и 0 - если нет

```python
#объединяю датафреймы и оставляю нужные столбцы
join_genre = filtered_films.merge(movies, on='movie_id', how='inner')[['movie_id', 'rating', 'release_year', 'unknown', 'Action', 'Adventure', 'Animation', 'Childrens', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']]  
join_genre.head()
```

  4.2. Добавьте к датафрейму столбцы с общим количеством оценок от всех пользователей на фильм и суммарной оценкой от всех пользователей

```python
#бъединяю датафреймы и оставляю нужные столбцы
join_ratig = join_genre.merge(ratigs_group, on='movie_id', how='inner')[['movie_id', 'rating', 'rating_count', 'rating_sum', 'release_year', 'unknown', 'Action', 'Adventure', 'Animation', 'Childrens', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']]  
join_ratig.head()
```

  5. Сформируйте X_train, X_test, y_train, y_test

```python
x = join_ratig[['rating_count', 'rating_sum', 'release_year', 'unknown', 'Action', 'Adventure', 'Animation', 'Childrens', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']]
y = join_ratig['rating']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
X_train.head()
```

  6. Возьмите модель линейной регрессии (или любую другую для задачи регрессии) и обучите ее на фильмах

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

  7. Оцените качество модели на X_test, y_test при помощи метрик для задачи регрессии

```python
print(round(model.score(X_test, y_test) * 100), '% точности на тестовых данных')
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test, model.predict(X_test))
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, model.predict(X_test))
```
